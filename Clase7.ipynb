### Generar una señal aleatoria y graficarla junto con su transformada Z.
### Se calcula la transformada Z de una señal discreta aleatoria utilizando la serie geométrica.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, exp
from scipy.stats import uniform

def transZfunc(x):
    Z = symbols('Z')
    N = len(x)
    X = sum(x[n]*Z**(-n) for n in range(N))
    return X

# Generamos la señal aleatoria
np.random.seed(12345)
x = uniform.rvs(size=10)

# Calculamos la transformada Z
Z = transZfunc(x)

# Evaluamos la transformada Z en puntos discretos
Z_eval = np.array([Z.subs('Z', np.exp(1j*w)) for w in np.linspace(-np.pi, np.pi, 100)], dtype=np.complex128)

# Graficamos la señal aleatoria y su transformada Z
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
plt.stem(x)
plt.title('Señal aleatoria')

plt.subplot(2, 1, 2)
plt.stem(np.abs(Z_eval))
plt.title('Transformada Z de la señal')

plt.tight_layout()
plt.show()
```
