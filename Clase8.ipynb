### En este ejemplo se calcula la transformada z de una señal aleatoria con una distribucion normal

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, exp

def transZfunc(x):
    Z = symbols('Z')
    N = len(x)
    X = sum(x[n]*Z**(-n) for n in range(N))
    return X

# Generar señal aleatoria
N = 1000  # Número de muestras
fs = 1000  # Frecuencia de muestreo en Hz
t = np.linspace(0, N/fs, N, endpoint=False)  # Vector de tiempo
x = np.random.normal(size=N)  # Señal aleatoria


# Calculamos la transformada Z
Z = transZfunc(x)

# Evaluamos la transformada Z en puntos discretos
Z_eval = np.array([Z.subs('Z', np.exp(1j*w)) for w in np.linspace(-np.pi, np.pi, 100)], dtype=np.complex128)

# Graficamos la señal aleatoria y su transformada Z
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
plt.stem(x)
plt.title('Señal aleatoria')

plt.subplot(2, 1, 2)
plt.stem(np.abs(Z_eval))
plt.title('Transformada Z de la señal')

plt.tight_layout()
plt.show()

```

### Para evaluar la expresión simbólica de la transformada $\mathbb{Z}$ en puntos discretos se usa el método subs() que sustituye todas las instancias de Z en la expresión simbólica por np.exp(1j*w), que es un valor numérico,el número complejo $e^{i\omega}$, que representa una frecuencia angular discreta en el rango $[-\pi, \pi]$.
