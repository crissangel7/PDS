### En este ejemplo se calcula la transformada z de una se√±al aleatoria con una distribucion normal

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, exp

def transZfunc(x):
    Z = symbols('Z')
    N = len(x)
    X = sum(x[n]*Z**(-n) for n in range(N))
    return X

# Generar se√±al aleatoria
N = 1000  # N√∫mero de muestras
fs = 1000  # Frecuencia de muestreo en Hz
t = np.linspace(0, N/fs, N, endpoint=False)  # Vector de tiempo
x = np.random.normal(size=N)  # Se√±al aleatoria


# Calculamos la transformada Z
Z = transZfunc(x)

# Evaluamos la transformada Z en puntos discretos
Z_eval = np.array([Z.subs('Z', np.exp(1j*w)) for w in np.linspace(-np.pi, np.pi, 100)], dtype=np.complex128)

# Graficamos la se√±al aleatoria y su transformada Z
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
plt.stem(x)
plt.title('Se√±al aleatoria')

plt.subplot(2, 1, 2)
plt.stem(np.abs(Z_eval))
plt.title('Transformada Z de la se√±al')

plt.tight_layout()
plt.show()

```

### Para evaluar la expresi√≥n simb√≥lica de la transformada $\mathbb{Z}$ en puntos discretos se usa el m√©todo subs() que sustituye todas las instancias de Z en la expresi√≥n simb√≥lica por np.exp(1j*w), que es un valor num√©rico,el n√∫mero complejo $e^{i\omega}$, que representa una frecuencia angular discreta en el rango $[-\pi, \pi]$.
