## Ejemplo clase 1


```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2*np.pi, 100) # Vector de tiempo con más muestras
signal = np.exp(-t)             # Señal analógica

w = int(input("Digite una frecuencia de muestreo: "))
x = np.linspace(0, 2*np.pi, w) # Nuevo vector de tiempo con W muestras
sampled_signal = np.exp(-x)     # Señal digitalizada

plt.figure()
plt.plot(t, signal, 'b', label = 'Señal Analógica')
plt.stem(x, sampled_signal, 'r', label = 'Señal Digitalizada')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()
```
