### Generar una señal exponencial compleja y graficarla

#### estamos definiendo una señal exponencial compleja, Combinacion Seno y Coseno.
#### Con amplitud 1, frecuencia 2 Hz, fase pi/4 radianes y frecuencia de muestreo de 100 Hz. 
#### Estamos generando un vector de tiempo discreto n con una duración de 1 segundo (100 muestras).

```python
import numpy as np
import matplotlib.pyplot as plt

# definimos los parámetros de la señal exponencial compleja
A = 1  # Amplitud
f = 2  # Frecuencia en Hz
phi = np.pi/4  # Fase en radianes
fs = 100  # Frecuencia de muestreo en Hz
T = 1/fs  # Periodo de muestreo
n = np.arange(0, fs)  # Vector de tiempo discreto

x = A * np.exp(1j * 2 * np.pi * f * n * T + phi)    # Generando la exponencial compleja

plt.stem(n, np.real(x), linefmt='blue', markerfmt='bo', label='Real')
plt.stem(n, np.imag(x), linefmt='red', markerfmt='ro', label='Imaginary')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.title('Funcion Exponencial Compleja')
plt.legend()
plt.show()
```
