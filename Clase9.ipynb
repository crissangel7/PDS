## Ejemplo Clase 9
## Hallar la transformada inversa de una funcion de trasferencia
### En este ejemplo, hemos utilizado impulse() para calcular la respuesta al impulso de la función de transferencia y graficarla. La salida es una gráfica de la respuesta al impulso de la función de transferencia en el dominio continuo.


```python
import numpy as np
from scipy import signal

# Definir la función de transferencia
b = [1, -0.5]
a = [1, 0.2]
sys = signal.TransferFunction(b, a)

# Calcular la transformada inversa
n = 50
t, y = signal.impulse(sys, N=n)

# Graficar la respuesta impulso
import matplotlib.pyplot as plt
plt.plot(t, y)
plt.xlabel('n')
plt.ylabel('h[n]')
plt.title('Respuesta al impulso de H(z) = (z - 0.5) / (z + 0.2)')
plt.show()
```

### Ejemplo 2
### Graficar la respuesta al impulso y el diagrama de Bode de una función de transferencia en el dominio Z
### Se utiliza la función bode para calcular la magnitud y la fase de la función de transferencia en un rango de frecuencias logarítmicas.

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Definir la función de transferencia
b = [1, -0.5]
a = [1, 0.2]
H = signal.TransferFunction(b, a)

# Calcular la respuesta al impulso
n = 50
t, h = signal.impulse(H, N=n)

# Graficar la respuesta al impulso
plt.stem(t, h)
plt.xlabel('n')
plt.ylabel('h[n]')
plt.title('Respuesta al impulso de H(z)')

# Graficar la función de transferencia
w, mag, phase = H.bode()
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.semilogx(w, mag)
ax1.set_ylabel('Magnitud (dB)')
ax1.set_title('Diagrama de Bode de H(z)')
ax2.semilogx(w, phase)
ax2.set_xlabel('Frecuencia (rad/s)')
ax2.set_ylabel('Fase (grados)')

plt.show()
```
