### Generar una señal aleatoria dedistribucion normal y aplicarle un filtro de media movil con la ventana del filtro ingresada por teclado.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar señal aleatoria
N = 1000  # Número de muestras
fs = 1000  # Frecuencia de muestreo en Hz
t = np.linspace(0, N/fs, N, endpoint=False)  # Vector de tiempo
x = np.random.normal(size=N)  # Señal aleatoria

# Aplicar filtro de media móvil
window_size = int(input("Digite el tamaño de la ventana del filtro: "))  # Tamaño de la ventana
window = np.ones(window_size)/float(window_size)
y = np.convolve(x, window, 'same')  # Señal filtrada

# Graficar señal original y filtrada
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.plot(t, x)
ax1.set_ylabel('Amplitude')
ax1.set_title('Original Signal')
ax2.plot(t, y)
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Amplitude')
ax2.set_title('Filtered Signal')
plt.show()
```

### la señal filtrada es más suave que la señal original, lo que indica que el filtro de media móvil está reduciendo el ruido de la señal.
### Mientras mayor sea el numero de la ventana mas suavisada va ser la grafica y va verse mejor pero va ser menos precisa en cuanto a la señal original.
